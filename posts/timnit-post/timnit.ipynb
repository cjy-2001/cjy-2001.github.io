### FATE in Computer Vision

In her 2020 talk at the Conference on Computer Vision and Pattern Recognition, Dr. Gebru discussed the importance of incorporating fairness, accountability, transparency, and ethics (FATE) in computer vision systems. She highlighted various examples where biased data, flawed algorithms, and lack of transparency have led to unintended consequences and discriminatory outcomes. Dr. Gebru emphasized the need for a multidisciplinary approach to addressing these issues, including involving social scientists, ethicists, and domain experts in the development and deployment of AI systems.

- Issues with Abstraction:

The practice of abstraction in Computer Vision and Machine Learning results in marginalized groups experiencing disparities in automated facial analysis tools. Both race and gender, as social constructs, have fluctuating classifications across time and space, leading to problems such as misgendering and harm to the transgender community.

- Dataset Prejudices:

Datasets often suffer from a lack of diversity, with a predominantly white and male representation. This bias extends to real-life scenarios, including medical research and car crash testing. Automation prejudice, when combined with surveillance, can lead to damaging consequences for vulnerable populations.

- Facial Recognition Tech and Insufficient Regulation:

Facial recognition technology exacerbates systemic racism and bias, as evidenced by its use in policing, racial profiling, and other oppressive methods. Companies like Faception and HireVue demonstrate the risks of amplifying existing biases. The AI field lacks proper safety testing and scrutiny, making it less regulated than industries such as pharmaceuticals.

- Self-protection:

People can counteract surveillance systems using fashion items, long hair, or infrared LEDs. The primary lesson is that Computer Vision today intensifies systemic racism and bias, causing tangible harm. It is crucial to address these concerns and view technology from a social standpoint to comprehend its far-reaching impact.

**tl;dr: As computer vision becomes increasingly integrated into our daily lives, it is crucial to ensure that these systems are designed and deployed responsibly, taking into account their potential biased outcomes.**
